```python
# mean evid score -- save 6
# 4 is answer-only
# 5 is concat
# 6 is evid
table = pd.DataFrame(columns=["save_num", "mean_ques_score", "mean_evid_score", "verdict_acc", "justi_score"])
for save_num in [4,5,6]:
    df = pd.read_json(
        f"/mnt/data/factcheck/averimatec/prepare_submission/intermediate_eval_results/gemma_gemma_{save_num}.json"
    )
    df
    (
        df["ques_score"].mean(),
        df["evid_score"].mean(),
        len(df[(df["evid_score"] > 0.3) & (df["verdict_score"] == 1.0)]) / len(df),
        ((df["evid_score"] > 0.3) * (df["justi_score"])).mean(),
        #(df["justi_score"]).mean(),
    )
    table = pd.concat(
        [
            table,
            pd.DataFrame(
                {
                    "save_num": [save_num],
                    "mean_ques_score": [df["ques_score"].mean()],
                    "mean_evid_score": [df["evid_score"].mean()],
                    "verdict_acc": [
                        len(df[(df["evid_score"] > 0.3) & (df["verdict_score"] == 1.0)])
                        / len(df)
                    ],
                    "justi_score": [((df["evid_score"] > 0.3) * (df["justi_score"])).mean()],
                }
            ),
        ],
        ignore_index=True,
    )
```
